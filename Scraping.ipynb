```python
import os
import requests

def download_folder(url, folder_name):
    # Create the folder if it doesn't exist
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    # Make a GET request to the GitHub API endpoint
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Loop through each file in the response
        for file in response.json():
            # Check if the file is a directory
            if file["type"] == "dir":
                # Recursively download the contents of the directory
                download_folder(file["url"], os.path.join(folder_name, file["name"]))
            # If the file is a file, download the file
            elif file["type"] == "file":
                # Make a GET request to the file URL
                file_response = requests.get(file["download_url"])
                # Save the content to a file with the same name in the specified folder
                with open(os.path.join(folder_name, file["name"]), "wb") as f:
                    f.write(file_response.content)
    else:
        print(f"Error retrieving contents of {url}: {response.status_code}")

# Call the function to download the folder contents
download_folder("https://api.github.com/repos/rmtonmoy/SpacyDatasetGenerator/contents/product_names", "product_names")
```

```python
import cv2
import pytesseract

# load the video
cap = cv2.VideoCapture(r"/Users/jashwanthreddy14/THD/Videos/Unclog_toilet.mp4")

# set the frame rate and get the first frame
fps = int(cap.get(cv2.CAP_PROP_FPS))

# set the start time to 0 milliseconds
start_time = 0

# set the end time to 30 seconds
end_time = 30 * 1000

# set the current time to the start time
current_time = start_time

# loop through the time range in 1-second intervals
while current_time < end_time:
    # set the position of the video capture to the current time
    cap.set(cv2.CAP_PROP_POS_MSEC, current_time)
    
    # read the current frame
    ret, frame = cap.read()
    
    # convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # apply thresholding to enhance the text
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    
    # perform text segmentation
    contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        roi = frame[y:y+h, x:x+w]
        text = pytesseract.image_to_string(roi, lang='eng')
        print(text)
        
    # update the current time to the next second
    current_time += 1000
```

```python
import cv2
import kraken

# load the video
cap = cv2.VideoCapture(r"/Users/jashwanthreddy14/THD/Videos/Unclog_toilet.mp4")

# set the frame rate and get the first frame
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
ret, frame = cap.read()

# loop through the frames
for i in range(min(20 * fps, frame_count)):
    # check only one frame per second
    if i % fps != 0:
        continue

    # extract the text from the frame
    text = kraken.recognize(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))

    # print the extracted text
    print(text)

    # get the next frame
    ret, frame = cap.read()
    if not ret:
        break

# release the video capture
cap.release()
```

```python
import cv2
import os

# Read the video
cap = cv2.VideoCapture(r"/Users/jashwanthreddy14/THD/Videos/Unclog_toilet.mp4")

# Create a directory to store the frames
if not os.path.exists("frames"):
    os.makedirs("frames")

# Set the starting point of the video (in seconds)
start_sec = 0

# Set the number of frames to extract
num_frames = 20

# Set the frame rate (in seconds)
frame_rate = 1

# Set the current frame
current_frame = 0

# Loop through the video
while cap.isOpened() and current_frame < num_frames:
    # Get the current time (in seconds)
    current_sec = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000

    # If we have reached the starting point
    if current_sec >= start_sec:
        # Read the frame
        ret, frame = cap.read()

        # If we have successfully read the frame
        if ret:
            # Save the frame
            cv2.imwrite(f"frames/frame{current_frame}.jpg", frame)

            # Increment the current frame
            current_frame += 1

            # Set the next frame to extract
            start_sec += frame_rate

    # Move to the next frame
    cap.set(cv2.CAP_PROP_POS_MSEC, (start_sec * 1000))

# Release the video capture
cap.release()
```

```python
import cv2

# Load video file
vidcap = cv2.VideoCapture(r"/Users/jashwanthreddy14/THD/Videos/Unclog_toilet.mp4")

# Get video duration and calculate frames per second (fps)
duration = int(vidcap.get(cv2.CAP_PROP_POS_MSEC))
fps = int(vidcap.get(cv2.CAP_PROP_FPS))

# Set interval to extract frames
interval = 1000  # 1 second
num_frames = 20

# Loop through the video and extract frames at set interval
frames = []
for i in range(num_frames):
    # Calculate frame index based on interval and fps
    frame_index = int((i * interval) / 1000 * fps)
    
    # Set the video frame to extract
    vidcap.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
    
    # Read the frame
    success, image = vidcap.read()
    
    # Check if frame was successfully read
    if success:
        # Convert the frame to grayscale and append it to the list of frames
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        frames.append(gray)
    else:
        print(f'Error extracting frame {i}')

# Release video capture object
vidcap.release()

# Display extracted frames
for i, frame in enumerate(frames):
    cv2.imshow(f'Frame {i}', frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2

video_file = r'/Users/jashwanthreddy14/THD/Videos/Unclog_toilet.mp4'
output_dir = r'/Users/jashwanthreddy14/THD/Images/'

# Open the video file
video = cv2.VideoCapture(video_file)

# Check if the video file was opened successfully
if not video.isOpened():
    print("Error: Could not open the video file.")
    exit()

# Get the frames per second (fps) of the video
fps = int(video.get(cv2.CAP_PROP_FPS))

# Loop through the video frames
frame_count = 0
while video.isOpened():
    # Read the next frame
    ret, frame = video.read()

    # Break the loop if we have reached the end of the video
    if not ret:
        break

    # Save the frame if it corresponds to a whole second (up to 20 seconds)
    if frame_count % fps == 0 and frame_count // fps <= 25:
        output_file = f"{output_dir}frame_{frame_count // fps}.png"
        cv2.imwrite(output_file, frame)

    frame_count += 1

# Release the video file
video.release()
```
